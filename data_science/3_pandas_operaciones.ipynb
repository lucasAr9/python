## Manejo de valores NaN

```python
# dropna para eliminar campos con NaN del DataFrame
netflix3 = netflix.dropna(axis=0, how='all')
# en el parametro axis= 0 si es fila; 1 si es columna;
# en el parametro how='all' si es para eliminar las filas o columnas que cumplan con la condicion
# en el parametro how='any' si es para eliminar todas las filas o columnas que cumplan con la condicion
netflix3.shape
```

```python
# para saber cuantos son los valores NaN en cada columna especificada
netflix3['duration'].isnull().values.ravel().sum()
# pd.isnull(netflix3['duration']).values.ravel().sum()
# values devuelve los valores de la columna, me quedo con los que son null y con sum() los sumo
```

```python
# para ver cuantas filas tienen NaN en cada columna
netflix3.isna().sum()
```

```python
# para quedarnos solo con los que sean NaN en la columna cast
los_nan = netflix['cast'].isnull()
los_nan
```

```python
# con fillna rellenar los valores NaN con un valor especificado en la columna seleccionada,
# en este caso, con la media (mean())
col = netflix3['release_year'].mean()
netflix3['release_year'].fillna(col, inplace=True)
netflix3.head(5)

# tambien en una sola columna en el mismo DataFrame
# netflix3['release_year'].fillna(80, inplace=True)

# tambien de la manera aplicandolo en todas las columnas
# netflix3.fillna(80, inplace=True)
```

```python
# para generar una nueva columna con datos calculados
netflix3['nueva_columna'] = col
netflix3.head(5)
```

## Concatenar dos o mas DataFrames

```python
# con iloc para seleccionar las filas que queremos
# los primeros 1:1 son las filas y la proxima 2:2 son las columnas dentro del iloc[]
# si se pone solo dos puntos (iloc[:, 1]) toma todas las filas de la columna 1
# si se pone solo dos puntos (iloc[1, :]) toma todas las columnas de la fila 1
# tambien se puede poner una lista con las filas o columnas que queremos
n1 = netflix3.iloc[7:12]
n2 = netflix3.iloc[0:5]
```

```python
# para concatenar dos DataFrames
n3 = pd.concat([n1, n2])
n3
```

## Concatenar dos o mas columnas en un DataFrame

```python
# para este ejemplo separamos tres columnas en net1 y dos en net2
net1 = netflix3[['type', 'title', 'country']]
net2 = netflix3[['duration', 'nueva_columna']]
```

## Merge de DataFrames

```python
# para obtener un dataset con todos los registros de los dos datasets
pd.merge(netflix, netflix2, how='outer', left_index=True, right_index=True)
```

```python
# para obtener un dataset con los registros en comun de los dos datasets
pd.merge(netflix, netflix2, how='inner', left_index=True, right_index=True)
```

```python
# para obtener un dataset con los registros del dataset izquierdo y con la informacion de las columnas
# correspondiente del dataset derecho, lo mismo pero al reves con 'right' en el parametro how
pd.merge(netflix, netflix2, how='left', left_index=True, right_index=True)
```

```python
# para obtener un dataset en donde se ubican los registros exactos segun la lista de indices
# detecta cuales son los nombres o valores que se repiten en las columnas seleccionadas (inner)
pd.merge(netflix, netflix2, how='inner', on=['title', 'release_year'])
```
